```python
import pandas as pd

from telco_customer_churn.config import RAW_DATA_DIR

df = pd.read_csv(RAW_DATA_DIR / "raw.csv")
```

```python
# "TotalCharges" should be numeric, but it's being treated as str (empty values?)
non_numeric_mask = ~df["TotalCharges"].str.replace(".", "", regex=False).str.isnumeric()
non_numerics = df[non_numeric_mask]["TotalCharges"]

percentage = len(non_numerics) / len(df)
print("Non-numeric TotalCharges values:", non_numerics.unique().tolist())
print(f"TotalCharges has {len(non_numerics)} non-numeric values ({percentage:.2%})")

df[df["TotalCharges"] == " "].head()
```

```python
# Convert "TotalCharges" to numeric, coercing errors to NaN, for further analysis
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```
