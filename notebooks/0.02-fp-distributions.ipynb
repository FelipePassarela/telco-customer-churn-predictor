```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from telco_customer_churn.config import RAW_DATA_DIR

df = pd.read_csv(RAW_DATA_DIR / "raw.csv")
```

```python
df = df.drop(columns=["customerID"])

# "TotalCharges" should be numeric, but it has some non-numeric values (white spaces strings).
# For plotting distributions, we need to convert it to numeric, coercing non-numeric values to NaN
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

categorical_cols = df.select_dtypes(include="str").columns.to_list()
numerical_cols = df.select_dtypes(exclude="str").columns.to_list()

print("Categorical columns:", *categorical_cols, sep="\n- ")
print("\nNumerical columns:", *numerical_cols, sep="\n- ")
```

```python
sns.set_theme(style="darkgrid", context="notebook")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
axes = axes.flatten()
for ax, col in zip(axes, numerical_cols):
    sns.histplot(data=df, x=col, ax=ax, kde=True, bins=30)
plt.tight_layout()
plt.show()
```

```python
# "SeniorCitizen" is a binary column (0 or 1), so it doesn't make sense to treat
# it as numerical. Let's convert it to categorical.

df["SeniorCitizen"] = df["SeniorCitizen"].map({0: "No", 1: "Yes"})
categorical_cols.append("SeniorCitizen")
numerical_cols.remove("SeniorCitizen")
```

```python
fig, axes = plt.subplots(1, len(numerical_cols), figsize=(9, 4))
for ax, col in zip(axes, numerical_cols):
    sns.violinplot(data=df, y=col, ax=ax)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, len(numerical_cols), figsize=(9, 4))
for ax, col in zip(axes, numerical_cols):
    sns.boxplot(data=df, y=col, ax=ax)
plt.tight_layout()
plt.show()
```

```python
target = "Churn"
sns.countplot(data=df, x=target, hue=target)
plt.title(f"{target} (target) distribution")
plt.tight_layout()
plt.show()
```
