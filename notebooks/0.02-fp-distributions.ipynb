```python
import pandas as pd

from telco_customer_churn.config import RAW_DATA_DIR

df = pd.read_csv(RAW_DATA_DIR / "raw.csv")
```

```python
df = df.drop(columns=["customerID"])

# "TotalCharges" should be numeric, but it has some non-numeric values (white spaces strings).
# For plotting distributions, we need to convert it to numeric, coercing non-numeric values to NaN
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

categorical_cols = df.select_dtypes(include="str").columns.to_list()
numerical_cols = df.select_dtypes(exclude="str").columns.to_list()

print("Categorical columns:", *categorical_cols, sep="\n- ")
print("\nNumerical columns:", *numerical_cols, sep="\n- ")
```

```python
import matplotlib.pyplot as plt

df.hist(bins=30, figsize=(9, 6), grid=False, edgecolor="black")
plt.tight_layout()
plt.show()
```

```python
df.plot(
    kind="box",
    subplots=True,
    figsize=(9, 4),
    grid=False,
    title="Box plots of numerical columns",
)
plt.tight_layout()
plt.show()
```

```python
target = "Churn"
df[target].value_counts().plot(
    kind="bar", figsize=(6, 4), grid=False, edgecolor="black"
)

plt.title(f"{target} (target) distribution")
plt.tight_layout()
plt.show()
```
